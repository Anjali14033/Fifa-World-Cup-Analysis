```python
data = data.replace('\?','',regex = True).astype(str) #It only replace the question mark ,'',in the word
data.head(10)
```

```python
import numpy as np
data = data.replace('not applicable',np.NaN ) #It replace the "not applicable" into "nan" in the column
data.head(10)
```

```python
data = data.dropna(subset= ["Give Name"],axis = 0)
data.head(10)
```

```python
data.shape
```

```python
data = data.replace('WC-','',regex = True).astype(str)
data.head(10)
```

```python
#Merging both column Given Name and Family Name together into single columns called Name
data['Name'] = data['Give Name'] + ' ' + data['Family Name']
data.drop(['Family Name', 'Give Name'], axis = 1, inplace = True)
data.head()
```

```python
#creating new column Year and will drop column Tournament Id
year_l=[]
for i in data["Tournament Name"]:
    year = int(i[0:4])
    year_l.append(year)

year_l=np.array(year_l)

data.insert(2,'Year',year_l)
data.head(5)
```

```python
data = data.drop("Tournament Id",axis =1)
data.head()
```

```python
##SEEING TYPE OF AWARDS IN DATA-SET
data["Award Name"].unique()
```

```python
#Single Line Chart
y = data["Award Name"].head(15)
x = data["Name"].head(15)
plt.xticks(rotation = 90)
plt.plot(x,y)

plt.title("Single Line Plot")
plt.ylabel("Award Name")
plt.xlabel("Name")
plt.grid(True)

plt.show()
```

```python
#Multiple Line Chart
x1 = data["Award Name"].head(20)
y1 = data["Name"].head(20)
#plt.xticks(rotation = 90)
x2 = data["Award Name"].head(30)
y2 = data["Tournament Name"].head(30)
plt.xticks(rotation = 90)

plt.plot(x1, y1, label='Line 1')
plt.plot(x2, y2, label='Line 2')

plt.title("Multiple Line Plot")
plt.ylabel("Name")
plt.xlabel("Award Name")
plt.grid(True)

plt.show()
```

##Visualising top 10 player who won max number of awards

```python
##seeing top 10 player who won max number of awards
players= data.groupby("Name").size().sort_values(ascending = False).head(10)

##visualising top 10 player who won max number of awards
players.plot(kind="bar",title="Player with highest award",ylabel ="No. of awards")
```

## CHECKING WHICH COUNTRY WON AWARD MAXIMUM NUMBER OF TIME

```python
## CHECKING WHICH COUNTRY WON AWARD MAXIMUM NUMBER OF TIME

data.groupby("Team Name").size().sort_values(ascending = False).head(10).plot(kind = "bar",title ="Country wise awards",ylabel="No. of awards")
```

## CHECKING WHICH COUNTRY WON AWARD MAXIMUM NUMBER OF TIME

```python
awards = data['Award Name'].value_counts()

plt.figure(figsize = (6,6))

awards.plot.pie(autopct = "%1.0f%%", colors = sns.color_palette('winter_r'), shadow = True)

c = plt.Circle((0,0), 0.4, color =  'white')
plt.gca().add_artist(c)
plt.title('Awards Percentage Type Won By Players')
plt.show()
```
